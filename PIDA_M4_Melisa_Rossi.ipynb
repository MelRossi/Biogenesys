### Análisis Estadístico

Se aplican estadísticas descriptivas con bucles.

Se decide dividir por grupos de variables según su relación. Por ejemplo: variables de población, variables según tasas de mortalidad, variables climáticas, etc.

```python
# Columnas a analizar
variables_analisis = [
    'new_confirmed', 
    'new_deceased', 
    'cumulative_confirmed',
    'cumulative_deceased', 
    'cumulative_vaccine_doses_administered'
]

# Agrupar por país
for pais, grupo in df_latinoamerica.groupby('country_name'):
    print(f"{pais}")
    
# Bucle for para esas columnas
    for columna in variables_analisis:
        print(f"Estadísticas para {columna}")
        media = grupo[columna].mean()
        desviacion_estandar = grupo[columna].std()
        minimo = grupo[columna].min()
        maximo = grupo[columna].max()

# Imprimir resultados
        print(f"  Media: {media:}")
        print(f"  Desviación Estándar: {desviacion_estandar:}")
        print(f"  Mínimo: {minimo:}")
        print(f"  Máximo: {maximo:}")

```

Función para estadísticas personalizadas. Se busca obtener la mediana, varianza y reango de las variables de nuestro dataset.

```python
# Función para estadísticas personalizadas

def estadisticas_personalizadas(serie):
    mediana = serie.median()
    varianza = serie.var()
    rango = serie.max() - serie.min()
    return {"Mediana": mediana, "Varianza": varianza, "Rango": rango}

# Uso de la función
for columna in df_latinoamerica.select_dtypes(include=['float64', 'int64']).columns:
    print(f"{columna}")
    resultado = estadisticas_personalizadas(df_latinoamerica[columna])
    print(f"  Mediana: {resultado['Mediana']}")
    print(f"  Varianza: {resultado['Varianza']}")
    print(f"  Rango: {resultado['Rango']}\n")
```

##### ***¿Qué representa la mediana?***
##### La mediana representa el valor central de una distribución ordenada, es decir, el punto que divide los datos en dos mitades iguales: el 50 % de los valores son menores y el 50 % mayores. La mediana es una medida de tendencia central más robusta frente a la media, en casos de mucha variabilidad, la mediana no se ve afectada por valores extremos.

##### ***¿Cómo varía la dispersión de los datos en el conjunto de datos analizado, en términos de la varianza y el rango?***
##### La varianza mide cuán dispersos están los valores respecto a la media, y el rango, muestra la diferencia entre el valor máximo y mínimo.
##### En variables como 'cumulative_confirmed', 'cumulative_vaccine_doses_administered', 'population', 'population_male', 'population_female', 'population_rural', 'population_urban', entre otras, presentan varianzas y rangos extremadamente altos, lo que demuestra una gran heterogeneidad en los datos.
##### En cambio, columnas como 'life_expectancy', 'comorbidity_mortality_rate', 'diabetes_prevalence', 'infant_mortality_rate' entre otras, tienen una varianza baja, lo que sugiere una distribución más estable y homogénea.

##### ***¿Qué nos puede indicar esto sobre la consistencia o la variabilidad de los datos en relación con la mediana?***
##### La comparación entre la mediana con la varianza y el rango (dispersión) permite interpretar la consistencia de los datos. 
##### En las columnas de acumulados y las referentes a población se observa una alta dispersión, por lo que los valores se ven alejados de la mediana.
##### En cambio en las columnas con baja varianza y rango (expectativa de vida, ratio de mortalidad infantil, comorbilidad, etc.), los valores tienden a ser más cercanos a la mediana.

```python
# Distribución de la población por grupos de edad

# Columnas a evaluar
columnas_edad = [
    'population_age_00_09',
    'population_age_10_19',
    'population_age_20_29',
    'population_age_30_39',
    'population_age_40_49',
    'population_age_50_59',
    'population_age_60_69',
    'population_age_70_79',
    'population_age_80_and_older'
]

# Agrupar por país y obtener la media
poblacion = df_latam.groupby('country_name')[columnas_edad].mean().reset_index()

# Convertir a formato largo 
poblacion_largo = pd.melt(
    poblacion,
    id_vars=['country_name'],
    value_vars=columnas_edad,
    var_name='Grupo de Edad',
    value_name='Población Media'
)

# Limpiar los nombres de las categorías de edad 
poblacion_largo['Grupo de Edad Limpio'] = (
    poblacion_largo['Grupo de Edad'].str.replace('population_age_', '').str.replace('_', '-')
)
poblacion_largo['Grupo de Edad Limpio'] = poblacion_largo['Grupo de Edad Limpio'].str.replace('-and-older', '+')


# Gráfico 
plt.figure(figsize=(16, 8))

sns.barplot(
    data=poblacion_largo,
    x='Grupo de Edad Limpio',
    y='Población Media',
    hue='country_name',
    palette=colores,
)

plt.title('Distribución de la Población (Media) por Grupos de Edad')
plt.ylabel('Población Media (Millones)')
plt.xlabel('Grupo de Edad')
plt.legend(title='País', loc='upper right')
plt.show()
```

```python
# Extraer el mes
df_latam['date'] = pd.to_datetime(df_latam['date'])
df_latam['mes'] = df_latam['date'].dt.to_period('M')

# Lista de paises
paises = ['Argentina', 'Brazil', 'Chile', 'Colombia', 'Mexico', 'Peru']

# Ajustar colores sún país
mapa_colores = dict(zip(paises, colores))

# Generar bucle y gráficos
for pais in paises:
    subset = df_latam[df_latam['country_name'] == pais]
    casos_mes = subset.groupby('mes')['new_confirmed'].sum().reset_index()
    casos_mes['mes'] = casos_mes['mes'].dt.to_timestamp()
    color_actual = mapa_colores.get(pais)
    
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=casos_mes, x='mes', y='new_confirmed', marker='o', color=color_actual)
    plt.title(f'Nuevos casos en {pais}')
    plt.ylabel('Nuevos casos')
    plt.xlabel('Mes del año')
    plt.ticklabel_format(style='plain', axis='y')
    plt.show()
```

Conclusión:

Patrones generales
- En todos los países analizados se observa un comportamiento estacional similar, con dos grandes picos de contagios: uno a mediados de 2021 y otro, mucho más marcado, a comienzos de 2022.

Comparación entre países
- Brasil y Argentina son los 2 países con mayor magnitud de casos. En Argentina, el pico de enero de 2022 supera los 2,9 millones de contagios mensuales, mientras que Brasil alcanza cerca de 3,4 millones.
- Chile y Perú presentan una curva más controlada durante la primera mitad de 2021, pero un aumento pronunciado a principios de 2022.
- Colombia y México muestran patrones ondulantes más marcados, con varios ciclos de repunte y descenso. 

Patrones temporales y estacionales
- Los picos se concentran principalmente entre los meses de diciembre y febrero, lo que podría sugerir una componente estacional (aunque no se encontró una relación importante entre la temperatura y cantidad de casos), o sociales (mayor movilidad y reuniones durante fin de año).

Muertes por mes de cada país

```python
# Generar bucle y gráficos
for pais in paises:
    subset = df_latam[df_latam['country_name'] == pais]
    muertes_mes = subset.groupby('mes')['new_deceased'].sum().reset_index()
    muertes_mes['mes'] = muertes_mes['mes'].dt.to_timestamp()
    color_actual = mapa_colores.get(pais)
    
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=muertes_mes, x='mes', y='new_deceased', marker='o', color=color_actual)
    plt.title(f'Muertes mensuales en {pais}')
    plt.ylabel('Nuevas Muertes')
    plt.xlabel('Mes del año')
    plt.show()
```

Conclusiones:

Patrones generales
- En todos los países analizados se observa una tendencia decreciente en el número de muertes a lo largo del tiempo.
Las cifras más elevadas se concentran durante la primera mitad de 2021, seguidas por un descenso sostenido en 2022, lo que refleja el impacto positivo de la vacunación y/o de una mejor respuesta sanitaria.

Comparación entre países
- Brasil es el país con mayor número absoluto de muertes, superando los 80.000 decesos mensuales en el pico de abril de 2021.
- México muestra una ola temprana a comienzos de 2021 (enero-febrero), con más de 35.000 muertes mensuales, seguida de un descenso notable y un rebrote más leve a mediados de ese año.
- Perú y Colombia presentan picos intensos al comienzo, alcanzando o superando los 20.000 fallecimientos mensuales.
- Argentina alcanza su punto máximo de mortalidad en mayo-junio de 2021, con cerca de 16.000 muertes.
- Chile, aunque presenta una menor magnitud de muertes en comparación, muestra un pico diferido en marzo de 2022, coincidente con el segundo brote.

Patrones temporales y estacionales
- La sincronía temporal entre países durante el primer semestre de 2021 sugiere una ola regional conjunta, reflejando dinámicas de contagio similares en América Latina.
- Durante 2022, los niveles de mortalidad se reducen drásticamente en todos los países, incluso a pesar de que los casos confirmados aumentaron (como se observó en el gráfico anterior).
Esto evidencia una desvinculación entre contagios y muertes, atribuible al efecto de la inmunización o mejores políticas sanitarias.

A diferencia de los casos confirmados, los picos de muertes no se repiten con igual intensidad en 2022.

Los resultados sugieren que la región experimentó un periodo crítico simultáneo, seguido por una recuperación sostenida que marca el inicio de la fase de control epidemiológico.

Evolución de casos confirmado vs muertes

```python
# Gráfico de líneas combinado

for pais in paises:
    subset = df_latam[df_latam['country_name'] == pais]
    
    # Agrupar datos
    resumen = subset.groupby('mes')[['new_confirmed', 'new_deceased']].sum().reset_index()
    resumen['mes'] = resumen['mes'].dt.to_timestamp()
    
    
    # Gráfico combinado
    plt.figure(figsize=(10, 4))
    sns.lineplot(data=resumen, x='mes', y='new_confirmed', label='Nuevos casos', color='#042940')
    sns.lineplot(data=resumen, x='mes', y='new_deceased', label='Nuevas muertes', color='#9FC131', linestyle='--')
    
    plt.title(f'Casos confirmados vs muertes - {pais}')
    plt.xlabel('Mes')
    plt.ylabel('Cantidad de personas')
    plt.legend()
    plt.xticks(rotation=45)
    plt.ticklabel_format(style='plain', axis='y')
    plt.show()
```

Conclusión general:

Encontramos que, en general, todos los paises presentan un patrón similar en el desarrollo de la pandemia. Se oobserva un crecimiento explosivo inicial, luego un período de control,un rebrote intenso a inicios de 2022, y finalmente una estabilización generalizada.

Tasa de crecimiento

```python
# Tasa de crecimiento

# Asugurar tipo de dato fecha
df_latam['date'] = pd.to_datetime(df_latam['date'])
df_latam = df_latam.sort_values(['country_name', 'date'])

# Agrupar semanalmente y calcular la tasa de crecimiento
crec_mensual = (
    df_latam
    .groupby(['country_name', pd.Grouper(key='date', freq='W')])['new_confirmed']
    .sum()
    .reset_index()
)

# Calcular tasa de crecimiento porcentual por país
crec_mensual['tasa_crecimiento'] = (
    crec_mensual
    .groupby('country_name')['new_confirmed']
    .pct_change() * 100
)

# Graficar
plt.figure(figsize=(16, 6))
sns.lineplot(
    data=crec_mensual,
    x='date', y='tasa_crecimiento', hue='country_name',
    marker='o', alpha=0.8,
    palette=colores
)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Tasa de crecimiento semanal de casos confirmados')
plt.xlabel('Semana')
plt.ylabel('Tasa de crecimiento (%)')
plt.tight_layout()
plt.show()
```

Conclusión:

- El pico inicial que observamos, se debe a que en las primeras semanas del registro los datos sobre nuevos casos era muy baja, por lo tanto, cualquier aumento se traduce en una tasa de crecimiento porcentual muy elevada.

- Luego del pico inicial, la tasa cae de manera pronunciada y se estabiliza cerca de valores bajos (entre cercanos al 20 o 30 %). Esto indica que hubo un menor crecimiento de nuevos casos.

- Durante el 2021 se observa una tasa de crecimiento estable, salvo por un pico de casos de Brasil entre las semanas de septiembre y octubre. 

- Finales de 2021 y comienzo de 2022 se observa un nuevo pico en la tasa de crecimiento que coincide en todos los países. Este nuevo brote puede sugerir una nueva variante de COVID.

- A partir de marzo pico las tasas bajan y se estabilizan nuevamente cerca de 0 %, indicando que el número de casos nuevos se mantuvo estable o en leve descenso.

- Volvemos a observar nuevos brotes, esta vez de manera más moderada, entre los meses de mayo y agosto de 2022.

Encontramos que, en general, todos los países presentan un patrón similar en el desarrollo de la pandemia. Se observa un crecimiento explosivo inicial, luego un período de control, un rebrote intenso a inicios de 2022, y finalmente una estabilización generalizada.
